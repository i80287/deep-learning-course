```python
def elu(x, alpha=0.05):
    pos_x = torch.relu(x)
    return pos_x + alpha * torch.expm1(x - pos_x)

def kaiming_init(n_in, n_out):
    return torch.normal(
        mean=torch.zeros(size=(n_in, n_out)),
        std=torch.full(size=(n_in, n_out), fill_value=np.sqrt(2 / n_in))
    ).to(device=DEVICE).requires_grad_(True)

def xavier_init(n_in: 'int', n_out: 'int') -> 'torch.Tensor':
    r = np.sqrt(6 / (n_in + n_out))
    return torch.FloatTensor(size=(n_in, n_out)).uniform_(-r, r).to(device=DEVICE).requires_grad_(True)

class CustomNet:
    def __init__(self, hidden_size=400, in_size=28*28, num_classes=n_classes):
        self.w1 = xavier_init(n_in=in_size, n_out=hidden_size)
        self.w2 = kaiming_init(n_in=hidden_size, n_out=hidden_size >> 1)
        self.w3 = xavier_init(n_in=hidden_size >> 1, n_out=num_classes)

        assert self.w1.requires_grad
        assert self.w2.requires_grad
        assert self.w3.requires_grad

    def mutable_parameters(self):
        return (self.w1, self.w2, self.w3)

    def train(self):
        for param in self.mutable_parameters():
            param.requires_grad_(True)

    def eval(self):
        for param in self.mutable_parameters():
            param.requires_grad_(False)

    def forward(self, x: 'torch.Tensor') -> 'torch.Tensor':
        batch_size, in_size = x.shape

        # (n x m) -> (batch_size x n x m)
        expand_weights = lambda w: w.unsqueeze(0).expand(batch_size, *w.shape)
        a = elu
        m = torch.bmm
        return a(m(
            a(m(
                a(m(                                # for vec : batch: (1 x in_size) @ (in_size x hidden_size) => (1 x hidden_size)
                    x.view(batch_size, 1, in_size), # (batch_size x in_size) -> (batchsize x 1 x in_size)
                    expand_weights(self.w1)         # (in_size x hidden_size) -> (batch_size x in_size x hidden_size)
                )),
                expand_weights(self.w2)             # (hidden_size x hidden_size) -> (batch_size x hidden_size x hidden_size)
            )),
            expand_weights(self.w3)                 # (hidden_size x num_classes) -> (batch_size x hidden_size x num_classes)
        )).squeeze_(1)
```

```python
net = CustomNet()
out = net.forward(torch.randn(2, 28*28, device=DEVICE))
assert len(out.shape) == 2
assert out.shape[-1] == n_classes
```

```python
import torch.nn.functional as F

def cross_entropy_loss(logits: 'torch.Tensor', target: 'torch.Tensor') -> 'torch.Tensor':
    N = logits.size(0)
    # Get the log probabilities
    log_probs = F.log_softmax(logits, dim=1)
    # Gather the log probabilities at the target indices
    log_probs_at_target = log_probs[torch.arange(log_probs.size(0)), target]
    # Compute the negative log likelihood
    nll = -log_probs_at_target.sum()
    return nll / N

y_tmp = torch.tensor(y_train[:2], device=DEVICE)
cross_entropy_loss(out, y_tmp), torch.nn.CrossEntropyLoss()(out, y_tmp)
```

```python
class CustomSGD:
    def __init__(self, model: CustomNet, lr=1e-4, l=0.0):
        assert 0 < lr <= 10
        assert 0 <= l < 1
        self.model = model
        self.lr = lr
        self.l = l

    def step(self):
        lr = self.lr
        with torch.no_grad():
            if (l := self.l):
                for param in self.model.mutable_parameters():
                    grad = param.grad
                    assert grad is not None
                    param.sub_(grad * (1 - l), alpha=lr)
            else:
                for param in self.model.mutable_parameters():
                    grad = param.grad
                    assert grad is not None
                    param.sub_(grad, alpha=lr)

    def zero_grad(self):
        for param in self.model.mutable_parameters():
            param.grad = None
```

```python
def iterate_minibatches(X, y, batch_size):
    indices = np.random.permutation(np.arange(len(X)))
    for start in range(0, len(indices), batch_size):
        ix = indices[start: start + batch_size]
        yield torch.from_numpy(X[ix]), torch.from_numpy(y[ix])
```

```python
def train(net: CustomNet, optimizer: CustomSGD, loss_fn=cross_entropy_loss, n_epoch=20):
    loss_history = []
    acc_history = []
    val_loss_history = []
    val_acc_history = []
    for _ in range(n_epoch):
        # Training
        net.train()
        acc_batches = []
        loss_batches = []
        for x_batch, y_batch in iterate_minibatches(X_train, y_train, batch_size=64):
            x_batch = x_batch.to(DEVICE)
            y_batch = y_batch.to(DEVICE)

            # Forward
            optimizer.zero_grad()
            out = net.forward(x_batch)

            loss = loss_fn(logits=out, target=y_batch)
            loss.backward()

            # Backward
            optimizer.step()

            # Accuracy
            loss_batches.append(loss.item())
            acc_batches += (out.argmax(axis=1) == y_batch).detach().cpu().tolist()

        loss_history.append(np.mean(loss_batches))
        acc_history.append(np.mean(acc_batches))

        # Validating
        net.eval()
        with torch.no_grad():
            acc_batches = []
            loss_batches = []
            for x_batch, y_batch in iterate_minibatches(X_val, y_val, batch_size=64):
                x_batch = x_batch.to(DEVICE)
                y_batch = y_batch.to(DEVICE)
                # Forward
                out = net.forward(x_batch)

                loss = loss_fn(logits=out, target=y_batch)
                # Accuracy
                loss_batches.append(loss.item())
                acc_batches += (out.argmax(axis=1) == y_batch).detach().cpu().tolist()

            val_loss_history.append(np.mean(loss_batches))
            val_acc_history.append(np.mean(acc_batches))

        clear_output(wait=True)
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
        ax1.set_xlabel("#epoch")
        ax1.set_ylabel("Loss")
        ax1.plot(loss_history, 'b', label='train loss')
        ax1.plot(val_loss_history, 'r', label='val loss')

        ax2.set_xlabel("#epoch")
        ax2.set_ylabel("Acc")
        ax2.plot(acc_history, 'b', label='train acc')
        ax2.plot(val_acc_history, 'r', label='val acc')
        plt.axhline(y = 0.8, color = 'g', linestyle = '--')

        plt.legend()
        plt.show()

    return max(val_acc_history)
```

```python
np.random.seed(5493284)
torch.manual_seed(0x82348372)
net = CustomNet(hidden_size=1100)
opt = CustomSGD(net, lr=0.005, l=0.15)
train(net, opt, cross_entropy_loss, n_epoch=70)
```

```python
def train_val_loop(net: torch.nn.Module, train_loader, val_loader, name: str, optimizer: torch.optim.Optimizer, criterion, n_epoch=20):
    loss_history = []
    acc_history = []
    val_loss_history = []
    val_acc_history = []

    for i in range(n_epoch):
        net.train()
        acc_batches = []
        loss_batches = []
        for x_batch, y_batch in train_loader:
            # Forward
            opt.zero_grad()
            out = net(x_batch.to(DEVICE))
            y_batch = y_batch.to(DEVICE)
            loss = criterion(out, y_batch)
            # Backward
            loss.backward()
            opt.step()
            # Accuracy
            loss_batches.append(loss.item())
            acc_batches += (out.argmax(axis=1) == y_batch).detach().cpu().tolist()

        loss_history.append(np.mean(loss_batches))
        acc_history.append(np.mean(acc_batches))

        # Validating
        net.eval()
        with torch.no_grad():
            acc_batches = []
            loss_batches = []
            for x_batch, y_batch in val_loader:
                # Forward
                out = net(x_batch.to(DEVICE))
                y_batch = y_batch.to(DEVICE)
                loss = criterion(out, y_batch)
                # Accuracy
                loss_batches.append(loss.item())
                acc_batches += (out.argmax(axis=1) == y_batch).detach().cpu().tolist()

            val_loss_history.append(np.mean(loss_batches))
            val_acc_history.append(np.mean(acc_batches))

        clear_output(wait=True)
        plt.figure(figsize=(8, 5))
        plt.title(f"Training/validating loss {name}")
        plt.xlabel("#epoch")
        plt.ylabel("Loss")
        plt.plot(loss_history, 'b', label='train')
        plt.plot(val_loss_history, 'r', label='validation')
        plt.legend()

        plt.figure(figsize=(8, 5))
        plt.title(f"Training/validating accuracy {name}")
        plt.xlabel("#epoch")
        plt.ylabel("Accuracy")
        plt.plot(acc_history, 'b', label='train')
        plt.plot(val_acc_history, 'r', label='validation')
        plt.legend()

        plt.show()

def test_accuracy(model):
    model.eval()
    test_acc_batches = []
    with torch.no_grad():
        for X_test, Y_test in test_loader:
            X_test = X_test.to(DEVICE)
            Y_test = Y_test.to(DEVICE)
            out = model.forward(X_test)
            test_acc_batches += (out.argmax(axis=1) == Y_test).detach().cpu().numpy().tolist()
    print(f'Test accuracy {np.mean(test_acc_batches)}')
```

```python
class TinyNeuralNetwork(nn.Module):
    def __init__(self, input_shape=28*28, num_classes=10, input_channels=1):
        super(self.__class__, self).__init__()
        hidden_size = 400
        self.model = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=input_shape * input_channels, out_features=hidden_size, bias=True),
            nn.ReLU(),
            nn.Linear(in_features=hidden_size, out_features=num_classes, bias=True),
        )

    def mutable_parameters(self):
        return self.model.parameters()

    def forward(self, inp):
        return self.model(inp)
```

## Task 3.3: Fix it.
Fix the overfitted network from the previous step (at least partially) by using regularization techniques (Dropout/Batchnorm/...) and demonstrate the results.

```python
class FixedNeuralNetwork(nn.Module):
    class SkipConn(nn.Module):
        def __init__(self, model: nn.Module):
            super().__init__()
            self.model = model

        def forward(self, x):
            return x + self.model(x)

    def __init__(self, input_shape=28*28, num_classes=10, input_channels=1):
        super().__init__()
        fc1 = nn.Linear(in_features=input_shape * input_channels, out_features=360, bias=True)
        fc2 = nn.Linear(in_features=fc1.out_features, out_features=900, bias=True)
        fc3 = nn.Linear(in_features=fc2.out_features, out_features=900, bias=True)
        fc4 = nn.Linear(in_features=fc3.out_features, out_features=900, bias=True)
        fc5 = nn.Linear(in_features=fc4.out_features, out_features=900, bias=True)
        fc6 = nn.Linear(in_features=fc5.out_features, out_features=num_classes, bias=True)
        torch.nn.init.kaiming_normal_(fc1.weight, nonlinearity="relu")
        torch.nn.init.kaiming_normal_(fc2.weight, nonlinearity="relu")
        torch.nn.init.kaiming_normal_(fc3.weight, nonlinearity="relu")
        torch.nn.init.kaiming_normal_(fc4.weight, nonlinearity="relu")
        torch.nn.init.kaiming_normal_(fc5.weight, nonlinearity="relu")
        self.model = nn.Sequential(
            nn.Flatten(),
            fc1,
            nn.BatchNorm1d(fc1.out_features),
            nn.ReLU(),
            nn.Dropout(0.6),
            fc2,
            nn.BatchNorm1d(fc2.out_features),
            nn.ReLU(),
            nn.Dropout(0.7),
            FixedNeuralNetwork.SkipConn(model=nn.Sequential(
                fc3,
                nn.BatchNorm1d(fc3.out_features),
                nn.ReLU(),
                nn.Dropout(0.7),
                FixedNeuralNetwork.SkipConn(model=nn.Sequential(
                    fc4,
                    nn.BatchNorm1d(fc4.out_features),
                    nn.ReLU(),
                    nn.Dropout(0.6),
                    FixedNeuralNetwork.SkipConn(model=nn.Sequential(
                        fc5,
                        nn.BatchNorm1d(fc5.out_features),
                        nn.ReLU(),
                        nn.Dropout(0.8),
                    )),
                )),
            )),
            fc6,
        )

    def mutable_parameters(self):
        return self.model.parameters()

    def forward(self, x):
        return self.model(x)
```

### Conclusions:
As can be seen, nn with lots of sequential linear (fully connected) layers separated only by activation functions (like ReLU, etc.) is likely to overfit (as can be seen on the plots for the "overfit model"). ### Linear transform layer
Also known as dense layer, fully-connected layer, FC-layer.
You should implement it.

- input:   **`batch_size x n_feats1`**
- output: **`batch_size x n_feats2`**

```python
class Linear(Module):
    """
    A module which applies a linear transformation
    A common name is fully-connected layer, InnerProductLayer in caffe.

    The module should work with 2D input of shape (n_samples, n_feature).
    """
    def __init__(self, n_in, n_out):
        super(Linear, self).__init__()

        # This is a nice initialization
        stdv = 1./np.sqrt(n_in)
        #it is important that we should multiply X @ W^T
        self.W = np.random.uniform(-stdv, stdv, size = (n_out, n_in))
        self.b = np.random.uniform(-stdv, stdv, size = n_out)

        self.gradW = np.zeros_like(self.W)
        self.gradb = np.zeros_like(self.b)

    def updateOutput(self, input):
        w = self.W.T
        b = self.b
        out = np.apply_along_axis(lambda x: x @ w + b, axis=1, arr=input)
        self.output = out
        return out

    def updateGradInput(self, input, gradOutput):
        w = self.W
        gi = np.apply_along_axis(lambda v: v @ w, axis=1, arr=gradOutput)
        self.gradInput = gi
        return gi

    def accGradParameters(self, input, gradOutput):
        self.gradW = np.einsum("ki,kj", gradOutput, input)
        self.gradb = gradOutput.sum(axis=0)

    def zeroGradParameters(self):
        self.gradW.fill(0)
        self.gradb.fill(0)

    def getParameters(self):
        return [self.W, self.b]

    def getGradParameters(self):
        return [self.gradW, self.gradb]

    def __repr__(self):
        s = self.W.shape
        q = 'Linear %d -> %d' %(s[1],s[0])
        return q
```

```python
def test_Linear():
    np.random.seed(42)
    torch.manual_seed(42)

    batch_size, n_in, n_out = 2, 3, 4
    for _ in range(100):
        # layers initialization
        torch_layer = torch.nn.Linear(n_in, n_out)
        custom_layer = Linear(n_in, n_out)
        custom_layer.W = torch_layer.weight.data.numpy()
        custom_layer.b = torch_layer.bias.data.numpy()

        layer_input = np.random.uniform(-10, 10, (batch_size, n_in)).astype(np.float32)
        next_layer_grad = np.random.uniform(-10, 10, (batch_size, n_out)).astype(np.float32)

        # 1. check layer output
        custom_layer_output = custom_layer.updateOutput(layer_input)
        layer_input_var = torch.from_numpy(layer_input).requires_grad_(True)
        torch_layer_output_var = torch_layer(layer_input_var)
        assert np.allclose(torch_layer_output_var.data.numpy(), custom_layer_output, atol=1e-6)

        # 2. check layer input grad
        custom_layer_grad = custom_layer.updateGradInput(layer_input, next_layer_grad)
        torch_layer_output_var.backward(torch.from_numpy(next_layer_grad))
        torch_layer_grad_var = layer_input_var.grad
        assert np.allclose(torch_layer_grad_var.data.numpy(), custom_layer_grad, atol=1e-6)

        # 3. check layer parameters grad
        custom_layer.accGradParameters(layer_input, next_layer_grad)
        weight_grad = custom_layer.gradW
        bias_grad = custom_layer.gradb
        torch_weight_grad = torch_layer.weight.grad.data.numpy()
        torch_bias_grad = torch_layer.bias.grad.data.numpy()
        assert np.allclose(torch_weight_grad, weight_grad, atol=1e-6)
        assert np.allclose(torch_bias_grad, bias_grad, atol=1e-6)
```
